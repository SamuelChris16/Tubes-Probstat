2. Peta Distribusi Responden Berdasarkan Domisili

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca file Excel
df = pd.read_excel('Dataset_Tubes.xlsx')

# Hitung frekuensi domisili (tanpa geospasial)
frekuensi = df['Domisili_saat_ini'].dropna().str.lower().value_counts().reset_index()
frekuensi.columns = ['Daerah', 'Frekuensi']

# --- Visualisasi 1: Bar Chart Semua Domisili ---
plt.figure(figsize=(12, max(6, len(frekuensi)*0.3)))  # tinggi dinamis
plt.barh(frekuensi['Daerah'][::-1], frekuensi['Frekuensi'][::-1], color='skyblue')
plt.xlabel('Jumlah Responden')
plt.ylabel('Daerah')
plt.title('Jumlah Responden Berdasarkan Domisili')
plt.tight_layout()
plt.show()

# --- (Opsional) Visualisasi 2: Pie Chart Semua Domisili ---
if len(frekuensi) <= 10:
    plt.figure(figsize=(8,8))
    plt.pie(frekuensi['Frekuensi'], labels=frekuensi['Daerah'], autopct='%1.1f%%', startangle=140)
    plt.title('Distribusi Domisili Responden')
    plt.tight_layout()
    plt.show()
```

5. Channel Layanan Favorit berdasarkan Layanan Perbankan

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data
df = pd.read_excel("Dataset_Tubes.xlsx")

# Ambil nama-nama kolom yang dibutuhkan
kolom_fitur = [
    'Manakah_channel_layanan_yang_lebih_Anda_sukai_ketika_Anda_akan_menggunakan_ﬁtur_ﬁtur_layanan_perbankan_berikut___Cek_Saldo',
    'Manakah_channel_layanan_yang_lebih_Anda_sukai_ketika_Anda_akan_menggunakan_ﬁtur_ﬁtur_layanan_perbankan_berikut___Mutasi_Rekening',
    'Manakah_channel_layanan_yang_lebih_Anda_sukai_ketika_Anda_akan_menggunakan_ﬁtur_ﬁtur_layanan_perbankan_berikut___Transfer_antar_rekening__sesama_bank',
    'Manakah_channel_layanan_yang_lebih_Anda_sukai_ketika_Anda_akan_menggunakan_ﬁtur_ﬁtur_layanan_perbankan_berikut___Transfer_antar_bank',
    'Manakah_channel_layanan_yang_lebih_Anda_sukai_ketika_Anda_akan_menggunakan_ﬁtur_ﬁtur_layanan_perbankan_berikut___Pembelian_Pulsa__eMoney_atau_voucher_voucher_lain',
    'Manakah_channel_layanan_yang_lebih_Anda_sukai_ketika_Anda_akan_menggunakan_ﬁtur_ﬁtur_layanan_perbankan_berikut___Pembayaran_Tagihan'
]

# Buat nama pendek untuk label fitur
nama_fitur = [
    'Cek Saldo',
    'Mutasi Rekening',
    'Transfer Sesama',
    'Transfer Antar Bank',
    'Pembelian Pulsa/eMoney/Voucher Lain',
    'Pembayaran Tagihan'
]

# Ubah format data ke bentuk long
df_long = df[kolom_fitur]
df_long.columns = nama_fitur  # Rename agar lebih singkat
df_melt = df_long.melt(var_name='Fitur', value_name='Channel')

# Buat tabel frekuensi
frekuensi = pd.crosstab(df_melt['Fitur'], df_melt['Channel'])

# Plot stacked bar chart horizontal
plt.figure(figsize=(10, 6))
bottom = [0] * len(frekuensi)
colors = plt.cm.Set2.colors  # ambil warna dari colormap

for i, channel in enumerate(frekuensi.columns):
    plt.barh(frekuensi.index, frekuensi[channel], left=bottom, label=channel, color=colors[i % len(colors)])
    bottom = [a + b for a, b in zip(bottom, frekuensi[channel])]

plt.xlabel('Jumlah Responden')
plt.title('Preferensi Channel Layanan Perbankan per Fitur (Stacked)')
plt.legend(title='Channel', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

6. Kategori Nominal Pinjaman Online berdasarkan Usia

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Baca data
df = pd.read_excel('Dataset_Tubes.xlsx')

# Ubah nama kolom (ganti sesuai nama asli)
df = df.rename(columns={
    'Berapa_biasanya_nominal_Anda_meminjam_di_pinjaman_online__fintech_lending': 'Pinjaman',
    'Usia__dalam_tahun': 'Usia'
})

# Filter data valid
df_filtered = df.dropna(subset=['Pinjaman', 'Usia'])

# Urutan kategori pinjaman (biar X axis tidak acak)
kategori_urut = ["< Rp 1.000.000,-", "Rp 1.000.000 - Rp 5.000.000,-", "Rp 5.000.000 - Rp 10.000.000,-", "> Rp 10.000.000,-"]
df_filtered['Pinjaman'] = pd.Categorical(df_filtered['Pinjaman'], categories=kategori_urut, ordered=True)

# Scatter plot
plt.figure(figsize=(10,6))
sns.stripplot(data=df_filtered, x='Pinjaman', y='Usia', jitter=True, alpha=0.6)

plt.title('Usia Responden untuk Tiap Kategori Nominal Pinjaman Online')
plt.xlabel('Kategori Nominal Pinjaman Online')
plt.ylabel('Usia')
plt.grid(True, axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

7. Distribusi Usia berdasarkan Jenis Kelamin

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dan siapkan data
df = pd.read_excel("Dataset_Tubes.xlsx")
df = df.rename(columns={
    'Usia__dalam_tahun': 'Usia',
    'Jenis_kelamin': 'Jenis_Kelamin'
})
df['Jenis_Kelamin'] = df['Jenis_Kelamin'].str.strip().str.lower()

# Mapping nama agar jadi 'male' dan 'female'
df['Jenis_Kelamin'] = df['Jenis_Kelamin'].replace({
    'laki-laki': 'male',
    'perempuan': 'female'
})

# Filter data valid
df_filtered = df.dropna(subset=['Usia', 'Jenis_Kelamin'])

# Set seaborn style agar lebih clean
sns.set(style="white")

# Buat plot
plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=df_filtered,
    x='Usia',
    hue='Jenis_Kelamin',
    fill=True,
    common_norm=False,
    alpha=0.6,
    linewidth=1.5,
    palette={'male': '#4C72B0', 'female': '#DD8452'}
)

# Atur label dan tampilan
plt.xlabel("Usia (tahun)", fontsize=12)
plt.ylabel("scaled density", fontsize=12)
plt.legend(title='gender', labels=['Laki-laki', 'Perempuan'], title_fontsize=12, fontsize=11)
plt.grid(False)
sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```

8. Distribusi Penghasilan per Bulan dengan Pendidikan Terakhir

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_excel("Dataset_Tubes.xlsx")

# Rename dan bersihkan
df = df.rename(columns={
    'Pendidikan_terakhir': 'Pendidikan_terakhir',
    'Penghasilan_per_bulan': 'Penghasilan_per_bulan'
})
df['Pendidikan_terakhir'] = df['Pendidikan_terakhir'].str.strip()
df['Penghasilan_per_bulan'] = df['Penghasilan_per_bulan'].str.strip()

# Drop NA
df_clean = df.dropna(subset=['Pendidikan_terakhir', 'Penghasilan_per_bulan'])

# Atur urutan kategori
df_clean['Pendidikan_terakhir'] = pd.Categorical(
    df_clean['Pendidikan_terakhir'],
    categories=['SD', 'SMP', 'SMA', 'D3', 'S1', 'S2', 'S3', 'Lainnya'],
    ordered=True
)

df_clean['Penghasilan_per_bulan'] = pd.Categorical(
    df_clean['Penghasilan_per_bulan'],
    categories=[
        '< Rp 1 juta',
        'Rp 1 juta – Rp 2 juta',
        'Rp 2 juta – Rp 5 juta',
        'Rp 5 juta – Rp 10 juta',
        'Rp 10 juta – Rp 20 juta',
        '> Rp 20 juta'
    ],
    ordered=True
)

# Buat palette gradasi biru terang ke gelap
blue_palette = sns.color_palette("Blues", desat=1, n_colors=6)

# Plot
plt.figure(figsize=(12, 6))
sns.countplot(
    data=df_clean,
    x='Pendidikan_terakhir',
    hue='Penghasilan_per_bulan',
    palette=blue_palette
)
plt.title('Grouped Bar Chart: Pendidikan Terakhir vs Penghasilan Bulanan')
plt.xlabel('Pendidikan Terakhir')
plt.ylabel('Jumlah Responden')
plt.xticks(rotation=0)
plt.legend(title='Penghasilan Bulanan', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

9. Tren Preferensi Pembayaran berdasarkan Usia

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dan siapkan data
df = pd.read_excel("Dataset_Tubes.xlsx")

# Ganti nama kolom agar lebih mudah
df = df.rename(columns={
    'Usia__dalam_tahun': 'Usia',
    'Manakah_dari_metode_berikut_yang_paling_sering_Anda_gunakan_untuk_melakukan_pembayaran_transaksi': 'Metode_Pembayaran'
})

# Bersihkan data
df_clean = df.dropna(subset=['Usia', 'Metode_Pembayaran'])
df_clean['Usia'] = pd.to_numeric(df_clean['Usia'], errors='coerce')

# Kelompokkan usia ke rentang
bins = [10, 20, 30, 40, 50, 60, 70, 80]
labels = ['10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
df_clean['Kelompok_Usia'] = pd.cut(df_clean['Usia'], bins=bins, labels=labels, right=False)

# Hitung frekuensi untuk setiap kombinasi usia & metode pembayaran
trend_data = df_clean.groupby(['Kelompok_Usia', 'Metode_Pembayaran']).size().reset_index(name='Jumlah')

# Visualisasi dengan line chart (ala Claus Wilke)
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=trend_data,
    x='Kelompok_Usia',
    y='Jumlah',
    hue='Metode_Pembayaran',
    marker='o'
)

plt.title('Tren Metode Pembayaran Berdasarkan Kelompok Usia')
plt.xlabel('Kelompok Usia')
plt.ylabel('Jumlah Responden')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Metode Pembayaran', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

10.Trend Tingkat Pemahaman tentang AI berdasarkan Pendidikan Terakhir

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_excel("Dataset_Tubes.xlsx")

# Bersihkan nama kolom
df = df.rename(columns={
    'Pendidikan_terakhir': 'Pendidikan_Terakhir',
    'Sejauh_mana_Anda_mengetahui_istilah__Kecerdasan_Buatan_atau_AI_atau_Artificial_Intelligence': 'Pemahaman_AI'
})

# Bersihkan nilai data
df['Pendidikan_Terakhir'] = df['Pendidikan_Terakhir'].astype(str).str.strip()
df['Pemahaman_AI'] = df['Pemahaman_AI'].astype(str).str.strip()

# Drop NaN
df_clean = df.dropna(subset=['Pendidikan_Terakhir', 'Pemahaman_AI'])

# Cek isi aktual kategori
print("Kategori Pendidikan:", df_clean['Pendidikan_Terakhir'].unique())
print("Kategori Pemahaman AI:", df_clean['Pemahaman_AI'].unique())

# Visualisasi tanpa preset kategori
plt.figure(figsize=(12, 6))
sns.countplot(
    data=df_clean,
    x='Pendidikan_Terakhir',
    hue='Pemahaman_AI',
    palette='viridis'
)
plt.title('Trend Tingkat Pemahaman AI Berdasarkan Pendidikan Terakhir')
plt.xlabel('Pendidikan Terakhir')
plt.ylabel('Jumlah Responden')
plt.xticks(rotation=30)
plt.legend(title='Tingkat Pemahaman AI', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

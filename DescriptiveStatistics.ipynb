Data Kuantitatif terhadap kolom data Usia

```python
import pandas as pd
from IPython.display import display

# Baca file Excel
df = pd.read_excel('Dataset_Tubes.xlsx')

# Ambil kolom usia
usia = df['Usia__dalam_tahun'].dropna()

# Hitung statistik deskriptif
mean = usia.mean()
q1 = usia.quantile(0.25)
median = usia.median()
q3 = usia.quantile(0.75)
p5 = usia.quantile(0.05)
p95 = usia.quantile(0.95)
std_dev = usia.std()
variance = usia.var()
minimum = usia.min()
maximum = usia.max()

# Buat DataFrame hasil statistik
stat_df = pd.DataFrame({
    "Statistik": [
        "Mean",
        "Q1 (25th percentile)",
        "Median (Q2)",
        "Q3 (75th percentile)",
        "P5 (5th percentile)",
        "P95 (95th percentile)",
        "Standard Deviation",
        "Variance",
        "Minimum",
        "Maximum"
    ],
    "Nilai": [
        f"{mean:.2f}",
        f"{q1:.2f}",
        f"{median:.2f}",
        f"{q3:.2f}",
        f"{p5:.2f}",
        f"{p95:.2f}",
        f"{std_dev:.2f}",
        f"{variance:.2f}",
        f"{minimum:.2f}",
        f"{maximum:.2f}"
    ]
})

print("=== Statistik Deskriptif: Usia (dalam tahun) ===")
display(stat_df)
```

Data kuantitatif untuk Modus dan Tabel Frekuensi pada Jenis Kelamin

```python
import pandas as pd
from IPython.display import display

# Baca file excel
df = pd.read_excel('Dataset_Tubes.xlsx')

# Ambil kolom jenis kelamin
jenis_kelamin = df['Jenis_kelamin'].dropna()

# Hitung banyak sampel (untuk perhitungan proporsi)
jumlah_sampel = len(jenis_kelamin)

# Perhitungan frekuensi dan proporsi
frekuensi = jenis_kelamin.value_counts()
proporsi = (frekuensi / jumlah_sampel * 100).round(4)

# Tabel Frekuensi & Proporsi
tabel = pd.DataFrame({
    'Frekuensi': frekuensi,
    'Proporsi (%)': proporsi
})

# Menghitung modus (kategori dengan frekuensi tertinggi)
modus = tabel['Frekuensi'].idxmax()

print("=== Tabel Frekuensi & Proporsi Jenis Kelamin ===")
display(tabel)
print(f"Modus: {modus}")
```
